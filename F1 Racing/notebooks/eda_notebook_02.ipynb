```python
df = df.sort_values(by=['year','round'], ascending=[True,True])
```

```python
df['driver_form_3_races'] = (
    df.groupby('driverId')['points']
    .transform(lambda x: x.rolling(window=3, min_periods=1).sum().shift(1))
)
```

```python
print(df[df['driverId'] == 1][['year', 'round', 'points', 'driver_form_3_races']].tail(10))
```

```python
# Group by the Team, calculate the 3-race rolling sum, and shift by 1
df['team_form_3_races'] = (
    df.groupby('constructorId')['points']
    .transform(lambda x: x.rolling(window=3, min_periods=1).sum().shift(1))
)
```

```python
df.grid
```

```python
df['grid'] = np.where(df['grid'] == 0, 20, df['grid'])
```

```python
df.info()
```

```python
columns_to_fill = ['driver_form_3_races', 'team_form_3_races']

df[columns_to_fill] = df[columns_to_fill].fillna(0)

print("Missing values in momentum columns:")
print(df[columns_to_fill].isnull().sum())
```

```python
df = pd.get_dummies(df, columns=['constructorRef'], drop_first=True)
```

```python
df
```

```python
train_df = df[df['year'] < 2023]
test_df = df[df['year'] >= 2023]

columns_to_drop = ['positionOrder', 'position', 'resultId', 'raceId', 'driverId', 'constructorId']

X_train = train_df.drop(columns=columns_to_drop, errors='ignore')
y_train = train_df['positionOrder']

X_test = test_df.drop(columns=columns_to_drop, errors='ignore')
y_test = test_df['positionOrder']

# 3. The Audit
print("Training Clues Shape (Past):", X_train.shape)
print("Testing Clues Shape (Future):", X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

cols_to_scale = ['year', 'round', 'grid', 'driver_form_3_races', 'team_form_3_races']

X_train[cols_to_scale] = scaler.fit_transform(X_train[cols_to_scale])
X_test[cols_to_scale] = scaler.transform(X_test[cols_to_scale])
```

```python
# X_train = X_train.drop(columns=['positionText'], errors='ignore')
# X_test = X_test.drop(columns=['positionText'], errors='ignore')

# Target the specific column and replace 'R' with the number 20
df['positionText'] = df['positionText'].replace('R', 20)

# You can also chain this to catch other text anomalies in that column!
df['positionText'] = df['positionText'].replace({'W': 20, 'D': 20})
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
import pandas as pd

# 1. The Senior Filter: Keep only math-safe columns
X_train_clean = X_train.select_dtypes(exclude=['object'])
X_test_clean = X_test.select_dtypes(exclude=['object'])

# 2. THE GLOBAL IMPUTER (CRITICAL FIX): Plug all remaining blank spaces with 0
X_train_clean = X_train_clean.fillna(0)
X_test_clean = X_test_clean.fillna(0)

# 3. Initialize the Model
f1_model = LinearRegression()

# 4. Train the AI (The Study Session)
print("Training the AI on historical data...")
f1_model.fit(X_train_clean, y_train)

# 5. Predict the Future (The Final Exam)
print("Predicting the 2023 season...")
predictions = f1_model.predict(X_test_clean)

# 6. Grade the AI's Performance
mae = mean_absolute_error(y_test, predictions)
r2 = r2_score(y_test, predictions)

print("\n--- The Results ---")
print(f"Mean Absolute Error (MAE): {mae:.2f} positions")
print(f"R-squared Score: {r2:.2f}")

# Peek at the first 5 predictions vs reality!
results_df = pd.DataFrame({'Actual Position': y_test.head(), 'Predicted Position': predictions[:5]})
print("\nFirst 5 Predictions vs Reality:")
print(results_df)
```
