```python
team_name_mapping = {
    'Royal Challengers Bengaluru': 'Royal Challengers Bangalore',
    'Punjab Kings': 'Kings XI Punjab',
    'Rising Pune Supergiants': 'Rising Pune Supergiant',
    'Delhi Daredevils': 'Delhi Capitals',
}

# Now apply to all columns
for col in ['batting_team', 'bowling_team', 'team1', 'team2', 'match_won_by', 'toss_winner']:
    grouped_df[col] = grouped_df[col].replace(team_name_mapping)
```
